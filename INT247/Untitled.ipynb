```python
from matplotlib import pyplot as plt
x = [5, 8, 10]
y = [12, 16, 6]
plt.bar(x,y,align='center')
plt.ylabel("Y axis")
plt.xlabel("X axis")
plt.title("Bar Graph")
plt.show()
```

```python
#Subplot()
import numpy as np
import matplotlib.pyplot as plt
x=np.arange(0,3*np.pi,0.1)
y_sin=np.sin(x)
y_cos=np.cos(x)
plt.subplot(2,1,1)#(nrows=no. of rows, ncols=no. of columns, index=index start from left goes to right)
plt.plot(x,y_sin)
plt.title("Sin Graph")
#spacing
plt.tight_layout(h_pad=2)#width=w_pad; parameters: h_pad:horizontal padding w_pad: width padding
plt.subplot(2,1,2)#
plt.plot(x,y_cos)
plt.title("Cos Graph")
#spacing
#plt.subplots_adjust(left=0.45,bottom=0.1,right=0.9,top=0.9)
plt.show()
```

```python
#example to show usage of xticks()
import matplotlib.pyplot as plt
x=[1,2,3,4,5,6]
y=[3,1,4,5,3,6]
labels=['A','B','C','D','E','F']
plt.plot(x,y)
plt.xlabel("X axis")
plt.ylabel("y axis")
plt.title("Line graph")
plt.xticks(x,labels)
plt.show()
```
