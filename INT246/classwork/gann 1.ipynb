```python
import numpy as np
def thresh(x):
    if int(x)>=0:
        return 1
    else:
        return 0
def create_ANN(x,y):
    il=x.shape[1]+1
    ol=1
    hl=0
    iw=np.random.uniform(low=-5,high=5,size=(il,))
    hw=0
    return iw 
def ann_output(x,y,iw):
    error=[]
    for i in range(x.shape[0]):
        n=iw[0]+np.dot(iw[1:],x[i,:])  
        o=thresh(n) 
        e=y[i]-o
        error.append(e)
    return error
def mse_error(error):
    mse=np.dot(error,error)/len(error)
    return mse
```

```python
import numpy as np

def create_pop(ps,pl):
    pop=np.random.uniform(low=-5,high=5,size=(ps,pl))
    return pop
def cal_fitness(pop,x,y):
    fitness=[]
    for i in range(pop.shape[0]):
        error=ann_output(x,y,pop[i])
        mse=mse_error(error)  
        fitness.append(1/mse)
    return fitness
def sel_parent(pop, fitness, num_parents):
    parents = np.empty((num_parents, pop.shape[1]))
    for parent_num in range(num_parents):
        max_fitness_idx = np.argmax(fitness)
        parents[parent_num, :] = pop[max_fitness_idx, :]
        fitness[max_fitness_idx] = -99999999999
    return parents
def crossover_parent(parents, offspring_size):
    offspring = np.empty(offspring_size)
    crossover_point = offspring_size[1]//2
    for k in range(offspring_size[0]):
        parent1_idx = k%parents.shape[0]
        parent2_idx = (k+1)%parents.shape[0]
        offspring[k, 0:crossover_point] = parents[parent1_idx, 0:crossover_point]
        offspring[k, crossover_point:] = parents[parent2_idx, crossover_point:]
    return offspring
def mutation_child(offspring_crossover, num_mutations=1):
    for i in range(num_mutations):
        pop_select=np.random.randint(0,offspring_crossover.shape[0])
        g_select=np.random.randint(0,offspring_crossover.shape[1])
        mut_value=np.random.uniform(low=-4.0, high=4.0, size=(1,1))
        offspring_crossover[pop_select,g_select]=mut_value
        print('Population ',pop_select,' Gene Selected ',g_select)
    return offspring_crossover
```

```python
#### OR GATE PROBLEM USING GA-ANN
import numpy as np
x=np.array([[0,0],[0,1],[1,0],[1,1]])
y=np.array([0,1,1,1])
print(x.shape)
weights=create_ANN(x,y)
print(weights)
errors=ann_output(x,y,weights)
print(errors)
mse=mse_error(errors)
print(mse)
############

############################  iteration with GA
print("Program start: ")
for i in range(1000):
    pop_size=8
    pop_length=len(weights)
    pop=create_pop(pop_size,pop_length)
    fitness=cal_fitness(pop,x,y)
    print('iteration ',i)
    if np.any(np.isinf(fitness)):
        break
    parents=sel_parent(pop,fitness,num_parents=4)
    child = crossover_parent(parents,offspring_size=(4, 3))
    child_mut=mutation_child(child,num_mutations=4)
    pop[0:parents.shape[0], :] = parents
    pop[parents.shape[0]:, :] = child_mut
fitness=cal_fitness(pop,x,y)
print(fitness)
w=pop[np.argmax(fitness)]
print('W0=',w[0])
print('W1=',w[1])
print('W2=',w[2])
print(ann_output(x,y,w))  
#####

```
