```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Identity Function
x=np.arange(0, 50)
y=x
plt.plot(x,y)
plt.xlabel("Input")
plt.ylabel("Activation ")
plt.title("Identity")
plt.show()
```

```python
# Binary Activation Function
x=np.arange(0, 50)
y=[]
for i in x:
    if i>=25:
        y.append(1)
    else:
        y.append(0)
plt.scatter(x,y)
plt.xlabel("Input")
plt.ylabel("Activation ")
plt.title("Binary")
plt.show()
```

```python
# Bipolar Activation Function
x=np.arange(0, 50)
y=[]
for i in x:
    if i>=25:
        y.append(1)
    else:
        y.append(-1)
plt.scatter(x,y)
plt.xlabel("Input")
plt.ylabel("Activation ")
plt.title("Bipolar")
plt.show()
```

```python
# Binary sigmoidal
x=np.arange(-10, 10)
y=1/(1+np.exp(-x))
print(y)
plt.scatter(x,y)
plt.xlabel("Input")
plt.ylabel("Activation ")
plt.title("Binary sigmoidal")
plt.show()
```

```python
# Bipolar sigmoidal
x=np.arange(-10, 10)
y=(1-np.exp(-x))/(1+np.exp(-x))
print(y)
plt.scatter(x,y)
plt.xlabel("Input")
plt.ylabel("Activation ")
plt.title("Bipolar sigmoidal")
plt.show()
```

```python
# ramp Activation
x=np.arange(-10,10)
y=[]
for i in x:
    if(i<=0):
        y.append(0)
    elif(0<i<5):
        y.append(i)
    else:
        y.append(5)

plt.plot(x,y)
plt.xlabel("Input")
plt.ylabel("Activation ")
plt.title("Ramp")
plt.show()    
```
