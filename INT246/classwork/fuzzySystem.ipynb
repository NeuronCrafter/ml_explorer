## Fuzzy expert system id for tip generation based service and quality

```python
import numpy as np
import skfuzzy as fuzzy
from skfuzzy import control as ctrl
import matplotlib.pyplot as plt
```

```python
service=ctrl.Antecedent(np.arange(1,11,1),'service')
service['low']=fuzzy.trimf(service.universe,[1,1,4])
service['medium']=fuzzy.trapmf(service.universe,[3,4,6,7])
service['high']=fuzzy.trapmf(service.universe,[6,8,10,10])
service.view()
plt.show()
```

```python
quality=ctrl.Antecedent(np.arange(1,11,1),'quality')
quality['low']=fuzzy.trimf(quality.universe,[1,1,4])
quality['medium']=fuzzy.trapmf(quality.universe,[3,4,6,7])
quality['high']=fuzzy.trapmf(quality.universe,[6,8,10,10])
quality.view()
plt.show()
```

```python
tip=ctrl.Consequent(np.arange(0,21),'tip')
tip.automf(3)
tip.view()
plt.show()
```

```python
rule1=ctrl.Rule(service['low'] & quality['low'], tip['poor'])
rule2=ctrl.Rule(service['medium'] & quality['medium'], tip['average'])
rule3=ctrl.Rule(service['high'] & quality['high'], tip['good'])
rule=[rule1,rule2,rule3]
```

```python
x=ctrl.ControlSystem(rule)
y=ctrl.ControlSystemSimulation(x)
```

```python
y.input['quality']=7
y.input['service']=8
y.compute()
print(y.output['tip'])
```

```python
tip.view(sim=y)
plt.show()
```
